```python
# 2. Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage
import skfuzzy as fuzz
from deap import base, creator, tools, algorithms
import random

# K-Medoids from pyclustering
from pyclustering.cluster.kmedoids import kmedoids

# 3. Load Dataset
df = pd.read_csv(r"/content/marketing_campaign.csv", sep="\t")

print("\n===== First 5 Rows =====")
print(df.head())
print("\n===== Last 5 Rows =====")
print(df.tail())

print("\n===== Dataset Info =====")
print(df.info())

print("\n===== Descriptive Statistics =====")
print(df.describe(include='all'))
print("Original Dataset Shape:", df.shape)
```

```python

# ========================
# 4. Data Cleaning
# ========================
df.drop(columns=['Z_CostContact', 'Z_Revenue', 'Complain'], errors='ignore', inplace=True)
df['Income'].fillna(df['Income'].median(), inplace=True)
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
df['Age'] = 2025 - df['Year_Birth']
df = df[(df['Age'] <= 100) & (df['Income'] < 600000)].copy()
df.reset_index(drop=True, inplace=True)

df['Marital_Status'] = df['Marital_Status'].replace({
    'Married': 'Partner', 'Together': 'Partner',
    'Single': 'Alone', 'Divorced': 'Alone', 'Widow': 'Alone',
    'Absurd': 'Alone', 'YOLO': 'Alone'
})

print(f"Shape after cleaning: {df.shape[0]} rows")

# ========================
# 5. Feature Engineering
# ========================
reference_date = pd.to_datetime("2025-12-15")
df['Tenure_Days'] = (reference_date - df['Dt_Customer']).dt.days
df['Children'] = df['Kidhome'] + df['Teenhome']
df['Is_Parent'] = (df['Children'] > 0).astype(int)

spending_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Total_Spending'] = df[spending_cols].sum(axis=1)
df['Premium_Products'] = df['MntWines'] + df['MntMeatProducts']

purchase_cols = ['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumDealsPurchases']
df['Total_Purchases'] = df[purchase_cols].sum(axis=1)

campaign_cols = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']
df['Total_Accepted_Campaigns'] = df[campaign_cols].sum(axis=1)

```

```python

# ========================
# 6. Deep Exploratory Data Analysis (EDA)
# ========================
sns.set(style="whitegrid", font_scale=1.1)

# Demographic Distributions
demographic_cols = ['Age', 'Income', 'Education', 'Marital_Status', 'Children']
for col in demographic_cols:
    plt.figure(figsize=(12, 5))
    if col in ['Age', 'Income']:
        plt.subplot(1, 2, 1)
        sns.histplot(df[col], bins=30, kde=True, color='skyblue')
        plt.title(f"{col} Distribution")
        plt.subplot(1, 2, 2)
        sns.boxplot(y=df[col], color='lightcoral')
        plt.title(f"{col} Boxplot")
    else:
        sns.countplot(y=df[col], order=df[col].value_counts().index, palette='viridis')
        plt.title(f"{col} Distribution")
    plt.tight_layout()
    plt.show()

# Spending Patterns by Category (Stacked Bars)
for category in ['Education', 'Marital_Status']:
    grouped = df.groupby(category)[spending_cols].mean()
    grouped.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20')
    plt.title(f"Average Spending by Product - Grouped by {category}")
    plt.ylabel("Average Spending")
    plt.xticks(rotation=45)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# Correlation Heatmap
key_numeric = ['Age', 'Income', 'Total_Spending', 'Total_Purchases',
               'Total_Accepted_Campaigns', 'Children', 'Recency', 'Tenure_Days']
corr = df[key_numeric].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0, square=True)
plt.title("Correlation Matrix of Key Features")
plt.tight_layout()
plt.show()

```

```python

# ========================
# 7. Preprocessing
# ========================
clustering_features = ['Age', 'Income', 'Recency', 'Tenure_Days',
                       'Total_Spending', 'Children', 'Total_Purchases',
                       'Total_Accepted_Campaigns', 'Is_Parent']
categorical_for_clustering = ['Education', 'Marital_Status']

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), clustering_features),
    ('cat', OneHotEncoder(drop='first', sparse_output=False), categorical_for_clustering)
])

X_preprocessed = preprocessor.fit_transform(df)
print(f"Preprocessed shape: {X_preprocessed.shape}")

```

```python


# ========================
# 8. Hierarchical Clustering
# ========================
sample_idx = np.random.choice(X_preprocessed.shape[0], min(500, X_preprocessed.shape[0]), replace=False)
linked = linkage(X_preprocessed[sample_idx], method='ward')
plt.figure(figsize=(12, 6))
dendrogram(linked, truncate_mode='level', p=5)
plt.title("Hierarchical Dendrogram")
plt.tight_layout()
plt.show()

```

```python

# ========================
# 9. Genetic Algorithm for Optimal k
# ========================
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

toolbox = base.Toolbox()
toolbox.register("attr_k", random.randint, 2, 8)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_k, n=1)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

def evaluate_k(individual):
    k = individual[0]
    if k < 2:
        return -1,
    try:
        # Random initial medoids
        initial_medoids = random.sample(range(len(X_preprocessed)), k)
        kmed_instance = kmedoids(X_preprocessed.tolist(), initial_medoids)
        kmed_instance.process()
        clusters = kmed_instance.get_clusters()
        labels = np.full(len(X_preprocessed), -1)
        for cid, cluster in enumerate(clusters):
            for idx in cluster:
                labels[idx] = cid
        score = silhouette_score(X_preprocessed, labels)
        return score,
    except:
        return -1,

toolbox.register("evaluate", evaluate_k)
toolbox.register("mate", tools.cxUniform, indpb=0.5)
toolbox.register("mutate", tools.mutUniformInt, low=2, up=8, indpb=0.2)
toolbox.register("select", tools.selTournament, tournsize=3)

random.seed(42)
pop = toolbox.population(n=12)
algorithms.eaSimple(pop, toolbox, cxpb=0.7, mutpb=0.3, ngen=8, verbose=False)
best = tools.selBest(pop, 1)[0]
k_optimal = best[0]
print(f"\nGA Optimal k: {k_optimal} (Silhouette: {best.fitness.values[0]:.3f})")

```

```python
# ========================
# 10. K-Medoids + Fuzzy Clustering
# ========================
# K-Medoids using pyclustering
initial_medoids = random.sample(range(len(X_preprocessed)), k_optimal)
kmed_instance = kmedoids(X_preprocessed.tolist(), initial_medoids)
kmed_instance.process()
clusters = kmed_instance.get_clusters()

labels_kmed = np.full(len(X_preprocessed), -1)
for cid, cluster in enumerate(clusters):
    for idx in cluster:
        labels_kmed[idx] = cid

df['Cluster_KMedoids'] = labels_kmed
sil_kmed = silhouette_score(X_preprocessed, labels_kmed)
print(f"K-Medoids (pyclustering) Silhouette: {sil_kmed:.3f}")

# Fuzzy C-Means
X_fuzzy = X_preprocessed.T
cntr, u, _, _, _, _, fpc = fuzz.cluster.cmeans(X_fuzzy, k_optimal, 2, error=0.005, maxiter=1000, seed=42)
df['Cluster_Fuzzy'] = np.argmax(u, axis=0)
sil_fuzzy = silhouette_score(X_preprocessed, df['Cluster_Fuzzy'])
print(f"Fuzzy Silhouette: {sil_fuzzy:.3f} | FPC: {fpc:.3f}")

```

```python

# ========================
# 11. Profiling & Marketing Insights
# ========================
profile_num = df.groupby('Cluster_KMedoids')[clustering_features].mean().round(2)
profile_cat = df.groupby('Cluster_KMedoids')[categorical_for_clustering].agg(lambda x: x.value_counts().idxmax())
profile = pd.concat([profile_num, profile_cat], axis=1)
print("\nFinal Cluster Profiles:")
display(profile)

print("\nMarketing Strategies:")
for c in range(k_optimal):
    row = profile.loc[c]
    size = len(df[df['Cluster_KMedoids'] == c])
    perc = size / len(df) * 100
    print(f"\nCluster {c} ({size} customers, {perc:.1f}%):")
    if row['Income'] > profile['Income'].mean() and row['Total_Spending'] > profile['Total_Spending'].mean():
        print(" → Premium: Exclusive offers, loyalty, upsell wines/meat")
    elif row['Children'] >= profile['Children'].mean() + 0.5:
        print(" → Family: Bundles, kid products, bulk deals")
    elif row['Total_Accepted_Campaigns'] < profile['Total_Accepted_Campaigns'].mean():
        print(" → Low Engagement: Reactivation, personalized discounts")
    else:
        print(" → Standard: General campaigns, cross-sell")

```

```python

# ========================
# 12. Visualizations
# ========================
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
sns.boxplot(x='Cluster_KMedoids', y='Total_Spending', data=df)
plt.title("Spending by Cluster")
plt.subplot(1,2,2)
sns.boxplot(x='Cluster_KMedoids', y='Income', data=df)
plt.title("Income by Cluster")
plt.tight_layout()
plt.show()

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_preprocessed)
plt.figure(figsize=(10,8))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=df['Cluster_KMedoids'], palette='deep', s=80, alpha=0.8)
plt.title(f"PCA Segments (Variance: {pca.explained_variance_ratio_.sum():.1%})")
plt.legend(title="Cluster")
plt.grid(alpha=0.3)
plt.show()

```
